#### Задание №1
Дано уравнение 
$$\varepsilon\frac{ds}{dt}+s=0,$$     $$t\in[0,1],$$     $$s(0)=1,$$    $$\varepsilon\ll1.$$

Необходимо найти:
1. Точное решение уравнения;
2. Построить график точного решения при $\varepsilon=0.1$ и $\varepsilon=0.01$;
3. Численно решить методом Эйлера уравнение для $\varepsilon=0.01$ с шагом $\Delta t=0.02$ и шагом $\Delta t=0.002$;
4. Построить графики для численных решений.

```python
# Пример вывода 2х графиков на одном рисунке
import numpy as np
import matplotlib.pyplot as plt
import math

from fontTools.ttLib.tables.T_S_I__1 import table_T_S_I__1
%matplotlib inline 
#графики встраиваются в блокнот;
%config InlineBackend.figure_format = 'svg' #векторный формат #векторный формат

def some_foo(t, e):
    # аналитическое решение
    return np.exp(-t/e)

e1 = 0.1    
e2 = 0.01

t = np.linspace(0, 1, 100)
s = [some_foo(i, e1) for i in t]
plt.plot(t, s, label='График при $\epsilon = 0.1$')

s2 = [some_foo(i, e2) for i in t]
plt.plot(t, s2, label='График при $\epsilon = 0.01$')

plt.grid(True)
plt.legend()
plt.xlabel('$t$')
plt.ylabel('$s(t)$')
plt.show()
```

```python
def euler_solve(t0, t1, y0, e, n):
    t, h = np.linspace(t0, t1, n, retstep=True)
    s = [y0]
    s_n = y0
    for i in t[1:]:
        s_n_1 = s_n/(1+h/e)
        s.append(s_n_1)
        s_n = s_n_1
    return t, s

t_0 = 0
t_1 = 1
s_0 = 1
e = 0.01
dt_1 = 0.02
dt_2 = 0.002
n1 = int((t_1 - t_0)/dt_1)
n2 = int((t_1 - t_0)/dt_2)

t1, s1 = euler_solve(t_0, t_1, s_0, e, n1)
plt.plot(t1, s1, label=f'График при $\epsilon = {e}$ и $dt = {dt_1}$', color = 'black')

t2, s2 = euler_solve(t_0, t_1, s_0, e, n2)
plt.plot(t2, s2, label=f'График при $\epsilon = {e}$ и $dt = {dt_2}$' , color = 'red')

plt.grid(True)
plt.legend()
plt.xlabel('$t$')
plt.ylabel('$s(t)$')
plt.show()
```

```python
#сравнение нормальных решений
e = 0.01

t_eu, s_eu = euler_solve(t_0, t_1, s_0, e, n1)
plt.plot(t_eu, s_eu, label=f'График при $\epsilon = {e} $', color = 'red')

t = np.linspace(0, 1, 500)
s = some_foo(t, e)
plt.plot(t, s, label=f'График при $\epsilon = {e}$', color = 'green')

plt.grid(True)
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
import numpy as np
from math import * #импортируем все функции

#константы для вычислений

#для перевода радиан в градусы
grad = pi/180 

g = 9.81 # -_-

# плотность выоздуха
ro = 1.240 

#скорость звука
a = 340

#закон сопротивления
def C_x(M):
    return 0.44
```

\begin{equation*}
\overrightarrow{Y} = 
\begin{pmatrix}
Y[0] \\
Y[1]\\
Y[2]\\
Y[3]
\end{pmatrix} 
 = 
\begin{pmatrix}
x \\
y \\
V_x \\
V_y
\end{pmatrix}
\end{equation*}


\begin{equation*}
\frac{d\overrightarrow{Y}}{dt} = \overrightarrow{f}(t,\overrightarrow{Y}) =
\begin{pmatrix}
f[0] \\
f[1] \\
f[2] \\
f[3]
\end{pmatrix} 
=
\begin{pmatrix}
\frac{dY[0]}{dt}\\
\frac{dY[1]}{dt}\\
\frac{dY[1]}{dt}\\
\frac{dY[1]}{dt}
\end{pmatrix}
=
\begin{pmatrix}
V_x \\
V_y \\
\frac{dV_x}{dt} \\
\frac{dV_y}{dt}
\end{pmatrix}
\end{equation*}

```python
#начальное положение м.т.
x_0 = 0 
y_0 = 0

#начальная скорость
V_0 = 100

#начальнчй угол
thetta_0 = 30

#масса
mass = 30

#площадь миделя
S_m = pi*0.1**2/4

#тяга
P = 0
```

```python
# Н.У
Y_0 = np.array([
    x_0,
    y_0,
    V_0*cos(thetta_0*grad),
    V_0*sin(thetta_0*grad),
    P*cos(thetta_0*grad),
    P*sin(thetta_0*grad)
])
Y_0
```

```python
# Переписать немного функцию правых частей dY/dt
def f(t,Y):
    x = Y[0]
    y = Y[1]
    V_x = Y[2]
    V_y = Y[3]
    P_x = Y[4]
    P_y = Y[5]

    theta = np.arctan(V_y/V_x)
    V = np.sqrt(V_x**2 + V_y**2)
    N = mass*g
    M = V/a
    F_drag = 0.5 * ro * V**2 * S_m * C_x(M)

    a_x = (-F_drag * np.cos(theta) - P_x)/mass
    a_y = (-F_drag * np.sin(theta) + P_y - N)/mass


    return np.array([
        V_x,
        V_y,
        a_x,
        a_y,
        0,
        0
    ])
```

```python
# переписать функцию под требования
def euler(t_0, t_1, y_0, dYdt, n_points):
    dt = (t_1 - t_0) / n_points
    t = np.linspace(t_0, t_1, n_points)
    y = np.zeros((n_points, len(y_0)))
    y[0] = y_0

    for i in range(1, n_points):
        # Вычисляем производные в текущей точке
        dydt = dYdt(t[i-1], y[i-1])
        # Метод Эйлера: y_{i+1} = y_i + dt * f(t_i, y_i)
        y[i] = y[i-1] + dt * dydt

    return y, t
    
```

```python
y,t = euler(0, 7, Y_0, f, 100)
```

```python
import scipy.integrate as integ

t0 = 0
t1 = 13
n = 1000

ODE = integ.ode(f)
ODE.set_integrator('dopri5')
ODE.set_initial_value(Y_0,t0)

t = np.linspace(t0,t1,n)
Y = []
Y.append(Y_0)
i = 0
y_current = Y_0[1]


while y_current >= -1e-8 and t[i] <= t1:
    ODE.integrate(t[i])
    y_solve = ODE.y
    Y.append(y_solve)
    y_current = y_solve[1]
    i += 1

Y = np.array(Y)
```
