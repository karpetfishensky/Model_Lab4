#### Задание №1
Дано уравнение 
$$\varepsilon\frac{ds}{dt}+s=0,$$     $$t\in[0,1],$$     $$s(0)=1,$$    $$\varepsilon\ll1.$$

Необходимо найти:
1. Точное решение уравнения;
2. Построить график точного решения при $\varepsilon=0.1$ и $\varepsilon=0.01$;
3. Численно решить методом Эйлера уравнение для $\varepsilon=0.01$ с шагом $\Delta t=0.02$ и шагом $\Delta t=0.002$;
4. Построить графики для численных решений.

```python
# Пример вывода 2х графиков на одном рисунке
import numpy as np
import matplotlib.pyplot as plt
import math

from fontTools.ttLib.tables.T_S_I__1 import table_T_S_I__1
%matplotlib inline 
#графики встраиваются в блокнот;
%config InlineBackend.figure_format = 'svg' #векторный формат #векторный формат

def some_foo(t, e):
    # аналитическое решение
    return np.exp(-t/e)

e1 = 0.1    
e2 = 0.01

t = np.linspace(0, 1, 100)
s = [some_foo(i, e1) for i in t]
plt.plot(t, s, label='График при $\epsilon = 0.1$')

s2 = [some_foo(i, e2) for i in t]
plt.plot(t, s2, label='График при $\epsilon = 0.01$')

plt.grid(True)
plt.legend()
plt.xlabel('$t$')
plt.ylabel('$s(t)$')
plt.show()
```

```python
def euler_solve(t0, t1, y0, e, n):
    t, h = np.linspace(t0, t1, n, retstep=True)
    s = [y0]
    s_n = y0
    for i in t[1:]:
        s_n_1 = s_n/(1+h/e)
        s.append(s_n_1)
        s_n = s_n_1
    return t, s

t_0 = 0
t_1 = 1
s_0 = 1
e = 0.01
dt_1 = 0.02
dt_2 = 0.002
n1 = int((t_1 - t_0)/dt_1)
n2 = int((t_1 - t_0)/dt_2)

t1, s1 = euler_solve(t_0, t_1, s_0, e, n1)
plt.plot(t1, s1, label=f'График при $\epsilon = {e}$ и $dt = {dt_1}$', color = 'black')

t2, s2 = euler_solve(t_0, t_1, s_0, e, n2)
plt.plot(t2, s2, label=f'График при $\epsilon = {e}$ и $dt = {dt_2}$' , color = 'red')

plt.grid(True)
plt.legend()
plt.xlabel('$t$')
plt.ylabel('$s(t)$')
plt.show()
```

```python
import numpy as np
from math import * #импортируем все функции

#константы для вычислений

#для перевода радиан в градусы
grad = pi/180 

g = 9.81 # -_-

# плотность выоздуха
ro = 1.240 

#скорость звука
a = 340

#закон сопротивления
def C_x(M):
    return 0.44
```

\begin{equation*}
\overrightarrow{Y} = 
\begin{pmatrix}
Y[0] \\
Y[1]\\
Y[2]\\
Y[3]
\end{pmatrix} 
 = 
\begin{pmatrix}
x \\
y \\
V_x \\
V_y
\end{pmatrix}
\end{equation*}


\begin{equation*}
\frac{d\overrightarrow{Y}}{dt} = \overrightarrow{f}(t,\overrightarrow{Y}) =
\begin{pmatrix}
f[0] \\
f[1] \\
f[2] \\
f[3]
\end{pmatrix} 
=
\begin{pmatrix}
\frac{dY[0]}{dt}\\
\frac{dY[1]}{dt}\\
\frac{dY[1]}{dt}\\
\frac{dY[1]}{dt}
\end{pmatrix}
=
\begin{pmatrix}
V_x \\
V_y \\
\frac{dV_x}{dt} \\
\frac{dV_y}{dt}
\end{pmatrix}
\end{equation*}

```python
#начальное положение м.т.
x_0 = 0 
y_0 = 0

#начальная скорость
V_0 = 100

#начальнчй угол
thetta_0 = 30

#масса
mass = 30

#площадь миделя
S_m = pi*0.1**2/4

#тяга
P = 0
```

```python
# Н.У
Y_0 = np.array([
    x_0,
    y_0,
    V_0*cos(thetta_0*grad),
    V_0*sin(thetta_0*grad),
    P*cos(thetta_0*grad),
    P*sin(thetta_0*grad)
])
Y_0
```

```python
# Переписать немного функцию правых частей dY/dt
def f(t,Y):
    x = Y[0]
    y = Y[1]
    V_x = Y[2]
    V_y = Y[3]
    P_x = Y[4]
    P_y = Y[5]

    theta = np.arctan(V_y/V_x)
    V = np.sqrt(V_x**2 + V_y**2)
    N = mass*g
    M = V/a
    F_drag = 0.5 * ro * V**2 * S_m * C_x(M)

    a_x = (-F_drag * np.cos(theta) - P_x)/mass
    a_y = (-F_drag * np.sin(theta) + P_y - N)/mass


    return np.array([
        V_x,
        V_y,
        a_x,
        a_y,
        0,
        0
    ])
```

```python
# переписать функцию под требования
def euler(t_0, t_1, y_0, dYdt, n_points):
    dt = (t_1 - t_0) / n_points
    t = np.linspace(t_0, t_1, n_points)
    y = np.zeros((n_points, len(y_0)))
    y[0] = y_0

    for i in range(1, n_points):
        # Вычисляем производные в текущей точке
        dydt = dYdt(t[i-1], y[i-1])
        # Метод Эйлера: y_{i+1} = y_i + dt * f(t_i, y_i)
        y[i] = y[i-1] + dt * dydt

    return y, t
    
```

```python
from scipy.integrate import ode
import matplotlib.pyplot as plt
import numpy as np

# ========== ПРОФИЛЬ ТЯГИ ==========
def thrust_profile(t):
    """Произвольный профиль тяги:
    - 0-1 сек: линейный рост до 200 Н
    - 1-3 сек: постоянная тяга 200 Н
    - 3-4 сек: линейный спад до 0 Н
    - после 4 сек: тяга 0 Н
    """
    if t < 1:
        return 200 * t  # линейный рост
    elif t < 3:
        return 200      # постоянная тяга
    elif t < 4:
        return 200 * (4 - t)  # линейный спад
    else:
        return 0        # выключение двигателя

# ========== ПРАВЫЕ ЧАСТИ С ТЯГОЙ ==========
def f_with_thrust(t, Y):
    x = Y[0]
    y = Y[1]
    V_x = Y[2]
    V_y = Y[3]

    # Вычисляем текущую тягу
    P = thrust_profile(t)
    theta = np.arctan2(V_y, V_x)  # направление тяги совпадает с направлением скорости

    P_x = P * np.cos(theta)
    P_y = P * np.sin(theta)

    V = np.sqrt(V_x**2 + V_y**2)
    N = mass * g
    M = V / a
    F_drag = 0.5 * ro * V**2 * S_m * C_x(M)

    a_x = (-F_drag * np.cos(theta) + P_x) / mass
    a_y = (-F_drag * np.sin(theta) + P_y - N) / mass

    return np.array([V_x, V_y, a_x, a_y])

# ========== УСЛОВИЕ ОСТАНОВКИ (ПАДЕНИЕ НА ЗЕМЛЮ) ==========
def ground_event(t, Y):
    """Событие - достижение земли (y <= 0)"""
    return Y[1]  # возвращаем y-координату

ground_event.terminal = True  # остановить интегрирование при событии
ground_event.direction = -1   # событие при переходе через 0 сверху вниз

# ========== РЕШЕНИЕ СИСТЕМЫ scipy.integrate.ode ==========
# Начальные условия (без компонент тяги)
Y0_scipy = np.array([
    x_0,
    y_0,
    V_0 * np.cos(thetta_0 * grad),
    V_0 * np.sin(thetta_0 * grad)
])

# Создаем решатель
solver = ode(f_with_thrust)
solver.set_integrator('dopri5')  # метод Дорманда-Принса 5-го порядка
solver.set_initial_value(Y0_scipy, 0)

# Интегрируем с автоматическим подбором шага
t_solution = [0]
Y_solution = [Y0_scipy]

while solver.successful() and solver.t < 30:  # максимальное время 30 сек
    solver.integrate(solver.t + 0.1)  # следующий шаг
    t_solution.append(solver.t)
    Y_solution.append(solver.y)

    # Проверяем условие падения
    if solver.y[1] <= 0:
        print(f"Объект упал на землю в t = {solver.t:.2f} сек")
        print(f"Координаты падения: x = {solver.y[0]:.2f} м, y = {solver.y[1]:.2f} м")
        break

Y_solution = np.array(Y_solution)

# ========== ГРАФИК ТРАЕКТОРИИ ==========
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(Y_solution[:, 0], Y_solution[:, 1], 'b-', linewidth=2, label='Траектория с тягой')
plt.plot(y[:, 0], y[:, 1], 'r--', alpha=0.7, label='Траектория без тяги')
plt.grid(True)
plt.xlabel('x, м', fontsize=12)
plt.ylabel('y, м', fontsize=12)
plt.title('Траектория полета')
plt.legend()
plt.axis('equal')

# Отметка точки падения
fall_idx = np.argmax(Y_solution[:, 1] <= 0)
if fall_idx > 0:
    plt.plot(Y_solution[fall_idx, 0], Y_solution[fall_idx, 1], 'ro', markersize=8, label='Точка падения')
    plt.legend()

# ========== ГРАФИК ПРОФИЛЯ ТЯГИ ==========
plt.subplot(1, 2, 2)
t_thrust = np.linspace(0, min(5, t_solution[-1]), 1000)
P_thrust = [thrust_profile(t) for t in t_thrust]

plt.plot(t_thrust, P_thrust, 'g-', linewidth=2)
plt.grid(True)
plt.xlabel('Время, сек', fontsize=12)
plt.ylabel('Тяга, Н', fontsize=12)
plt.title('Профиль тяги от времени')
plt.fill_between(t_thrust, P_thrust, alpha=0.3, color='green')

plt.tight_layout()
plt.show()

# ========== ВЫВОД РЕЗУЛЬТАТОВ ==========
print("\n=== РЕЗУЛЬТАТЫ ИНТЕГРИРОВАНИЯ ===")
print(f"Метод: DOPRI5 (scipy.integrate.ode)")
print(f"Время полета: {t_solution[-1]:.2f} сек")
print(f"Дальность полета: {Y_solution[-1, 0]:.2f} м")
print(f"Максимальная высота: {np.max(Y_solution[:, 1]):.2f} м")

# Сравнение с методом Эйлера
print(f"\n=== СРАВНЕНИЕ С МЕТОДОМ ЭЙЛЕРА ===")
print(f"Метод Эйлера - время полета: {t[-1]:.2f} сек")
print(f"Метод Эйлера - дальность: {y[-1, 0]:.2f} м")
print(f"Метод Эйлера - макс. высота: {np.max(y[:, 1]):.2f} м")

# ========== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ==========
# Вычисление скорости в момент падения
V_final = np.sqrt(Y_solution[-1, 2]**2 + Y_solution[-1, 3]**2)
print(f"\nСкорость в момент падения: {V_final:.2f} м/с")

# Анализ влияния тяги
print(f"\n=== ВЛИЯНИЕ ТЯГИ ===")
print(f"Максимальная тяга: {max(P_thrust):.0f} Н")
print(f"Общий импульс тяги: {np.trapz(P_thrust, t_thrust):.0f} Н·с")
```
